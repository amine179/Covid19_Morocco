# Here are the eight districts of Casablanca:

```python
import folium
import pandas as pd
import os
from folium import plugins

global m

states = os.path.join('CasaData', 'casa_districts.json')

m = folium.Map(location=[33.573109, -7.589843], zoom_start=12)

folium.GeoJson(
    states,
    name='Limites'
).add_to(m)

# Create markers
def mark(tool, x, y):
    global m
    folium.Marker([x, y],
                 popup='<strong>Welcome</strong>',
                 tooltip=tool,
                 icon=folium.Icon(color='green')).add_to(m)

mark('Casa anfa', 33.594150, -7.646590)   

mark('Hay Hassani', 33.545401,-7.688151)

mark('Ain Chock', 33.539678,-7.622030)

mark('Mers Sultan', 33.575154,-7.601243)


mark('Hay Mohamadi ain sbaa', 33.588740,-7.582496)

mark('Bernoussi sidi moumen', 33.612046,-7.492682)

mark('Ben msick Sidi Othman', 33.549121,-7.581989)

mark('Hay My Rrachid', 33.564284,-7.545239)


# To Add legend
legend_html =   '''
                <div style="position: fixed; 
                            bottom: 50px; left: 50px; width: 100px; height: 90px; 
                            border:2px solid grey; z-index:9999; font-size:14px;
                            ">&nbsp; <br>
                              &nbsp; Mettez pointeur sur : &nbsp; <i class="fa fa-map-marker fa-2x" style="color:green"></i><br>
                             
                </div>
                ''' 

m.get_root().html.add_child(folium.Element(legend_html))

plugins.Fullscreen(position='topright').add_to(m)

folium.LayerControl().add_to(m)


display(m)

```

```python
import folium
import pandas as pd
import os
from folium import plugins

global m

states = os.path.join('CasaData', 'casa_districts.json')

m = folium.Map(location=[33.573109, -7.589843], zoom_start=12)

folium.GeoJson(
    states,
    name='Limites'
).add_to(m)

# Create marker for each 3amala
def mark(tool, x, y, pop):
    global m
    folium.Marker([x, y],
                 popup=pop,
                 tooltip=tool,
                 icon=folium.Icon(color='green')).add_to(m)

mark('Casa anfa', 33.594150, -7.646590, 'centre ville, medina, maarif..')   

mark('Hay Hassani', 33.545401,-7.688151, 'zone populaire')

mark('Ain Chock sidi maarouf', 33.539678,-7.622030, 'Casa nearshore + zone industrielle')

mark('Mers Sultan', 33.575154,-7.601243, 'zone populaire avec les plus grands centre commerciaux populaires mais frms pour le moment ')


mark('Hay Mohamadi ain sbaa', 33.588740,-7.582496, 'zone industrielle')

mark('Bernoussi sidi moumen', 33.612046,-7.492682, 'zone industrielle')

mark('Ben msick Sidi Othman', 33.549121,-7.581989, 'que des quartiers populaires')

mark('Hay My Rrachid', 33.564284,-7.545239, 'que des quartiers populaires')


# add ant path route to map
#plugins.AntPath(points).add_to(m)

# To Add legend
legend_html =   '''
                <div style="position: fixed; 
                            bottom: 20px; left: 50px; width: 100px; height: 120px; 
                            border:4px solid grey; z-index:9999; font-size:14px;
                            ">&nbsp; <br>
                              &nbsp; Mettez pointeur ou clicker sur : &nbsp; <i class="fa fa-map-marker fa-2x" style="color:green"></i><br>
                             
                </div>
                ''' 

m.get_root().html.add_child(folium.Element(legend_html))

folium.raster_layers.TileLayer('cartodbpositron').add_to(m)
folium.raster_layers.TileLayer('Open Street Map').add_to(m)

data_rate = os.path.join('CasaData', 'casarate.csv')
state_data = pd.read_csv(data_rate)

m.choropleth(
    geo_data=states,
    name='taux d infection',
    data=state_data,
    columns=['Districts', 'infect_rate'],
    key_on='feature.id',
    fill_color='PuRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Infection rate'
)

# add full screen button to map
plugins.Fullscreen(position='topright').add_to(m)

folium.LayerControl().add_to(m)

display(m)
```
